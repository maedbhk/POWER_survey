```python
import os
import pandas as pd
import numpy as np
import plotly.express as px
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
# get data

fname = 'POWER_survey_2023.csv'
fdir = '/Users/maedbhking/Documents/POWER_survey/data'

df = pd.read_csv(os.path.join(fdir, fname))
```

## Who answered the survey?

```python
# of people answering the survey
print(f'{df.shape[0]} people answered the survey')

# professions
col_name = 'What is your job description at MIT?'
tmp = df[[col_name]].value_counts().reset_index()
ax = sns.barplot(x=col_name, y='count', data=tmp)
plt.xticks(rotation=45, ha='right');
plt.xlabel('')
plt.ylabel('# of respondents')
plt.title(col_name)

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

## Which departments are they from?

```python
def remap(x):
    remap = {'Brain and Cognitive Sciences/McGovern': 'McGovern',
             'picower\\brain and cognitive science': 'Picower',
             'Picower institute ': 'Picower',
             'McGovern': 'McGovern',
             'Brain and Cognitive Sciences': 'BCS',
             'Picower Institue for Learning and Memory': 'Picower',
             'McGovern Inst of Brain Research': 'McGovern',
             'cee': 'CEE',
             'Civil and Environmental Engineer': 'CEE', 
             'Cee': 'CEE', 
             'Civil and Environmental Engineering': 'CEE', 
             'Biological engineering/MIT': 'BE',
             'Department of biological engineering': 'BE',
             'Biological Engineering': 'BE',
             'physics': 'Physics',
             'chemistry': 'Chemistry',
             'Chemistry ': 'Chemistry',
             'Department of Chemistry': 'Chemistry',
             'ChemE': 'Chemical Engineering',
             'MIT Media Lab': 'Media Lab',
             'CSAIL ': 'CSAIL',
             'MIT Kavli Institute': 'Kavli',
             'MTI Kavli Institute': 'Kavli',
             'KI': 'Kavli'
            }
    if x in remap.keys():
        return remap[x]
    else:
        return x
    

df['dept'] = df['What is your department/institute?'].apply(lambda x: remap(x))
print(f'Respondents come from {len(df.dept.unique())} unique departments')

# departments - piechart

# define colors
colors = sns.color_palette('bright')[0:30]

tmp = df['dept'].value_counts().reset_index()
tmp.loc[tmp["count"] <= 2, "dept"] = 'other'
tmp1 = tmp.groupby('dept').apply(lambda x: x['count'].sum()).reset_index(name='count')

data = tmp1['count'].tolist()
labels = tmp1['dept'].tolist()

# create pie chart
plt.pie(data, labels=labels, colors=colors, autopct='%.0f%%')
plt.show()
```

## Length of parental leave

```python

def remap(x):
    remap = {'I am on leave now and taking the full 20 weeks': '20 weeks',
            'I was only offered 12 weeks of leave. I did not know that I could take 20 weeks, and would have done this if this were communicated to me.': '12 weeks',
            '16 -20 weeks': '16 weeks',
            '4 month paid, then unpaid break from working for 8 month and I am still not back to work': '16 weeks',
            'one semester': '12 weeks',
            '1 took 4 weeks': '4 weeks',
            '3 weeks ': '3 weeks',
            '4 months continuous, plus one month working 3 days a week': '16 weeks',
            '4 month': '16 weeks',
            '6 months': '24 weeks',
            '20': '20 weeks',
            'Took 4 weeks, another 8 in summer': '12 weeks',
            'I would like to take at least 4 weeks': '4 weeks',
            'I took 6 months of parental leave, but not while being at MIT. For a second child, I would plan for 9-12 months to avoid stress': '24 weeks',
             'Was not at MIT or the US at that time (9 months)': '36 weeks',
            'I would like to take 6 months, but the MIT policies forbid this.': '24 weeks',
            'Parental leave for 4 weeks': '4 weeks',
            '2 months': '8 weeks',
            '1 month': '4 weeks',
            'No': '0 weeks',
            'I took 10 weeks = 8 wks maternity leave and 2 wks parental leave': '10 weeks',
            '24weeks': '24 weeks',
            '12': '12 weeks',
            '3 months': '12 weeks',
            'Three months': '12 weeks',
            '12 weeks or more if allowed ': '12 weeks'
            }

    if x in remap.keys():
        return remap[x]
    else:
        return "NaN"
    
df['parental_leave'] = df['If you took parental leave: how long?\nIf you plan on taking parental leave: how long?'].apply(lambda x: remap(x))
df['parental_leave'] = df['parental_leave'].str.replace(' weeks', '')
df['parental_leave'] = df[df['parental_leave']!='NaN']['parental_leave'].astype(int)

col_name='parental_leave'
tmp = df[[col_name]].value_counts().reset_index()
tmp = tmp[tmp[col_name]!='Not applicable']
ax = sns.barplot(x=col_name, y='count', data=tmp)
plt.xticks(rotation=45, ha='right');
plt.xlabel('# of weeks')
plt.ylabel('# of respondents')
plt.title('If you took parental leave: how long? Or if you plan on taking parental leave: how long?')

for container in ax.containers:
    ax.bar_label(container, fmt='%.0f')
```

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS


def make_word_cloud(sentences, mask, output_file_name, stopwords=None, extra_stopwords=None,
                    bckgrd_color = "white", max_words=2000):
    
    """Generate a word cloud and write it to .png file.

    Keyword arguments:
    input_text -- path to plain text file
    mask -- path to mask image
    output_file_name -- string name to output .png file
    stopwords -- list of stop word strings (default None)
    extra_stopwords -- list of extra stop word strings (default None)
    bckgrd_color -- background color (default "white")
    max_words -- maximum number of words (default 2000)
    """
    
    sentences = df['What policies could MIT adopt that would make your life as a parent or parent-to-be easier?'].unique()
    text = ''.join(sentences[1:])

    # transform the text into a list of words
    words_list = text.split(' ')

    if mask is not None:
        # Load the mask image
        mask = np.array(Image.open(mask))
    # Load stop word list
    stopwords = set(stopwords)
    # Add extra stop words if provided
    if extra_stopwords is not None:
        [stopwords.add(word) for word in extra_stopwords]
    # Call WordCloud
    wc = WordCloud(background_color=bckgrd_color, max_words=max_words,
                   mask=mask, stopwords=stopwords)

    # Generate word cloud
    wc.generate(text)
    # Write to file
    wc.to_file(output_file_name)
```
